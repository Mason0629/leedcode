```python
#链表类
class LinkList():
    def __init__(self):
        #链表头 --> head结点，head是一个特殊的结节，head结点永远指向第一个结点
        self.head = None
        #链表尾 --> tail结点，tail结点也是一个特殊的结点，tail结点永远指向最后一个节点
        self.tail = None
        
    def is_empty(self):
        return self.head is None
    
    #尾部添加节点
    def append(self, data):
        #创建新的节点
        node = Node(data)
        #如果头尾或是尾为空 分别设置头尾
        if self.head is None:
            self.head = node
            self.tail = node
        else:
            #否则在为节点的添加node
            self.tail.next = node
            #然后重新定义为节点
            self.tail = node
    
    #遍历表链      
    def iter(self):
        #如果表头为空则为空表链
        if not self.head:
            print("LinkList is empty")
            return
        #定义指针，初始位置为表头
        cur = self.head
        yield cur.data
        #如果指针的next不为空，则意味链表并未到头
        while cur.next:
            cur = cur.next
            yield cur.data
    
    #链表插入
    def insert(self, idx, value):
        #绑定当前节点为表头节点
        cur = self.head
        # 初始化其实索引位置0
        cur_idx = 0
        if cur is None:
            raise Exception('The linklist is empty')
        #循环终止条件为 到达插入节点索引的前一个索引
        while cur_idx < idx-1:
            cur = cur.next
            if cur is None:
                raise Exception('List length is less than index')
            cur_index += 1
        #创建新的节点
        node = Node(value)
        #新节点的next指向当前节点的next
        node.next = cur.next
        #然后把当前节点的next指向新的node
        cur.next = node
        #如果是在最后一个节点插入，要变更尾节点
        if node.next is None:
            self.tail = node
            
    def remove(self, idx):
        cur = self.head
        cur_idx = 0
        if cur is None:
            raise Exception('The linklist is empty')
        while cur_idx < idx-1:
            cur = cur.next
            if cur is None:
                raise Exception('list length is less than index')
            cur_idx += 1
        if idx == 0:
            self.head = cur.next
            cur = cur.next
            return
        if self.head == self.tail:
            self.head = None
            self.tail = None
            return
        cur.next = cur.next.next
        if cur.next is None:
            self.tail = cur
        return
    
    def size(self):
        count = 0
        cur = self.head
        if cur is None:
            raise Exception('The linklist is empty')
        while cur:
            count += 1
            cur = cur.next
        return count
    
    def search(self, item):
        cur = self.head
        found = False
        while cur and not found:
            if cur.data == item:
                found = True
            else:
                cur = cur.next
        return found
```
